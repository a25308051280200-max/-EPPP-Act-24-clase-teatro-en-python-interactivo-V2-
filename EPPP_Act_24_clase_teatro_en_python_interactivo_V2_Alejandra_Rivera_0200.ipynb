```python
# Definición de la Clase
class Teatro:
    def __init__(self, nombre, aforo, tipoEscenario, historiaAnos):
        self.nombre = nombre
        self.aforo = aforo
        self.tipoEscenario = tipoEscenario
        self.historiaAnos = historiaAnos

    # --- Métodos (Funciones solicitadas) ---
    def ensayarObra(self):
        print(f"\n🎭 Iniciando ensayos en el '{self.nombre}'.")
        print(f"El elenco está utilizando el escenario de tipo: {self.tipoEscenario}.")

    def venderEntradas(self):
        print(f"\n🎫 Taquilla abierta en {self.nombre}.")
        print(f"Capacidad máxima disponible: {self.aforo} asientos.")

    def montarEscenografia(self):
        print(f"\n🔨 Preparando el escenario para la función.")
        print(f"Manteniendo la tradición de este recinto con {self.historiaAnos} años de historia.")

# --- Función Principal ---
def sistema_teatral():
    teatros = [] # Lista para guardar los dos teatros

    # Registro de los 2 teatros
    for i in range(1, 3):
        print(f"\n--- REGISTRO TEATRO {i} ---")
        nom = input(f"Nombre del teatro {i}: ")
        afo = input(f"Capacidad (Aforo) de {nom}: ")
        tipo = input(f"Tipo de escenario (ej. Proscenio, Circular): ")
        hist = input(f"Años de historia de {nom}: ")

        # Creamos el objeto y lo guardamos en la lista
        nuevo_teatro = Teatro(nom, afo, tipo, hist)
        teatros.append(nuevo_teatro)

    # Menú Interactivo
    while True:
        print("\n" + "═"*35)
        print("       MENÚ DE GESTIÓN TEATRAL")
        print("═"*35)
        print(f"1. Gestionar {teatros[0].nombre}")
        print(f"2. Gestionar {teatros[1].nombre}")
        print("3. Salir")

        opc_teatro = input("\n¿Qué teatro desea gestionar? (1-3): ")

        if opc_teatro in ["1", "2"]:
            # Elegimos el teatro según el índice (0 o 1)
            t = teatros[int(opc_teatro) - 1]

            print(f"\n--- Operaciones en: {t.nombre} ---")
            print("a) Ensayar Obra")
            print("b) Vender Entradas")
            print("c) Montar Escenografía")
            accion = input("Selecciona acción (a, b, c): ").lower()

            if accion == "a": t.ensayarObra()
            elif accion == "b": t.venderEntradas()
            elif accion == "c": t.montarEscenografia()
            else: print("Acción no válida.")

        elif opc_teatro == "3":
            print("\nCerrando sistema de gestión...")
            break
        else:
            print("\n[!] Opción no válida.")

# --- Ejecución ---
if __name__ == "__main__":
    sistema_teatral()

    print("\n" + "═"*45)
    print("Programa terminado con éxito")
    print("Desarrollado siguiendo el modelo de Alejandra Rivera Valenzuela")
    print("═"*45)
```
